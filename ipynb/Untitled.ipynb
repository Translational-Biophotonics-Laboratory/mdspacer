```python
%matplotlib inline

import matplotlib
import matplotlib.pyplot as plt

import numpy as np
import cv2
```

```python
def BGR2RGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
images = ['../s_C002Z016.tif', '../stitched(Direct File Access Mode)_Z067.tif']


for i in images:   
    img = cv2.imread(i, -1)
    scaled_img = cv2.normalize(img, dst=None, alpha=0, beta=255, norm_type=cv2.NORM_MINMAX)

    plt.figure(figsize=(15,15))
    plt.imshow(BGR2RGB(scaled_img))
```
