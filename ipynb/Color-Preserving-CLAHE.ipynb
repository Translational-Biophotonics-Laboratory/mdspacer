```python
%matplotlib inline

import os
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
def BGR2RGB(image):
    return cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
path = "C:\\Users\\danie\\Downloads\\fundo.png"

image = cv2.imread(path)

plt.figure(figsize=(5,5))
plt.axis('off')
plt.imshow(BGR2RGB(image))
```

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.figure(figsize=(5,5))
plt.axis('off')
plt.imshow(BGR2RGB(gray_image))
```

```python
# CLAHE on gray image
clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
clahe_image = clahe.apply(gray_image)

plt.figure(figsize=(5,5))
plt.axis('off')
plt.imshow(BGR2RGB(clahe_image))
```

```python
# CLAHE on color image
lab_image = cv2.cvtColor(image, cv2.COLOR_BGR2LAB)

clahe = cv2.createCLAHE(clipLimit=3.0, tileGridSize=(8,8))
lab_image[:,:,0] = clahe.apply(lab_image[:,:,0])

bgr_image = cv2.cvtColor(lab_image, cv2.COLOR_LAB2BGR)

plt.figure(figsize=(25,25))
plt.axis('off')
plt.imshow(BGR2RGB(bgr_image))
```
